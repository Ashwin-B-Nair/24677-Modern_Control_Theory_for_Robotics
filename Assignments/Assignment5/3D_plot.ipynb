```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def V_dot(x1, x2):
    return -4*x1**4*x2**2 
```

```python
x1 = np.linspace(-5, 5, 50)
x2 = np.linspace(-5, 5, 50)
X1, X2 = np.meshgrid(x1, x2)

vdot = V_dot(X1,X2)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(X1, X2, vdot, cmap='viridis')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('V_dot')
ax.set_title('3D plot of V_dot(x1, x2)')

plt.colorbar(surface)
plt.show()
```
