```python
import os
import numpy as np
import matplotlib.pyplot as plt
from control.phaseplot import phase_plot
from numpy import pi

plt.close('all')
```

```python
def org_system(x, t):
    # x1, x2 = X
    dx1dt = x[1] - x[0] * x[1]**2
    dx2dt = -x[1]
    return np.array([dx1dt, dx2dt])
```

```python
plt.figure(dpi = 300)
plt.clf()
plt.gca().set_aspect('equal', adjustable='box')
plt.axis([-1.5, 1.5, -1, 0.5])

plt.title('Phase Plot of Original System')
x1 = np.linspace(-1.5, 1.5, 5)
x2 = np.linspace(-1, 0.5, 5)
X1, X2 = np.meshgrid(x1, x2)
X0 = np.hstack((X1.reshape(-1,1), X2.reshape(-1,1)))

# Outer trajectories
phase_plot(
    org_system,
    X0 = X0,
    X = (-1.5, 1.5, 15),
    Y = (-1, 0.5, 15),
    T = 10
)

plt.scatter(x1 = [0],
            x2 = [0],
            c='r',
            alpha=0.5)

plt.show()
```

```python
def lin_system(X):
    x1, x2 = X
    dx1dt = x2
    dx2dt = x1*0
    return np.array([dx1dt, dx2dt])
```

```python
x1 = np.linspace(-5, 5, 50)
x2 = np.linspace(-5, 5, 50)
X1, X2 = np.meshgrid(x1, x2)
DX1, DX2 = lin_system([X1, X2])
```

```python
plt.figure(figsize=(8, 6))
plt.quiver(X1, X2, DX1, DX2, pivot='tip')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Phase Portrait of Linearized System')
plt.grid(True)
plt.axis('equal')
plt.show()
```
